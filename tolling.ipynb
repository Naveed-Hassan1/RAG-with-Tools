```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain.tools import tool
wkp = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())

@tool 
def add_number(a:int ,b:int)->int:
    "Adds two integers"
    return (int(a) + int(b))

@tool 
def multiply_number(a:int ,b:int)->int:
    "Multiply two integers"
    return (int(a) * int(b))

lis = [wkp,add_number,multiply_number]

llm_with_tool = llm.bind_tools(tools=lis)

llm_with_tool.invoke("Who is Imran Khan Niazi").tool_calls
```

```python
dic= {i.name:i for i in lis}
```

```python
from langchain_core.messages import HumanMessage, ToolMessage

Query = "What is sum of 10 and 20"

final = [HumanMessage(content=Query)]

res = llm_with_tool.invoke(Query)
final.append(res)

for tool_call in res.tool_calls:
    name = tool_call["name"]
    tool = dic[name] 
    try:
        result = tool.invoke(tool_call["args"])
    except Exception as e:
        result = f"Error: {str(e)}"
    final.append(ToolMessage(content=str(result), tool_call_id=tool_call["id"]))

print(final)
```
